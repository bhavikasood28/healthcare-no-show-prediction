```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Clean style
sns.set_theme(style="whitegrid")

# Bigger fonts for visuals
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['figure.dpi'] = 550  # HD resolution

df = pd.read_csv("../data/processed/processed_data.csv")
```

```python
df['target'].mean()
```

```python
plt.figure()
sns.barplot(data=df, x='days_wait', y='target', color='#1f77b4')
plt.title("No-Show Rate by Days Wait", fontweight='bold')
plt.xlabel("Days Between Scheduling & Appointment")
plt.ylabel("No-Show Probability")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.barplot(data=df, x='age_group', y='target', palette='Blues')
plt.title("No-Show Rate by Age Group", fontweight='bold')
plt.xlabel("Age Group")
plt.ylabel("No-Show Probability")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.barplot(data=df, x='sms_received', y='target', palette='viridis')
plt.title("Impact of SMS Reminders", fontweight='bold')
plt.xlabel("SMS Received (1 = Yes)")
plt.ylabel("No-Show Probability")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.barplot(data=df, x='appointment_day_of_week', y='target', palette='mako')
plt.title("No-Show Rate by Day of Week", fontweight='bold')
plt.xlabel("Day of Week (0 = Monday, 6 = Sunday)")
plt.ylabel("No-Show Probability")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x='patient_total_appointments',
    y='patient_past_no_shows',
    hue='target',
    palette='coolwarm'
)
plt.title("Patient Past Behavior vs No-Show", fontweight='bold')
plt.xlabel("Total Past Appointments")
plt.ylabel("Past No-Shows")
plt.tight_layout()
plt.show()
```
